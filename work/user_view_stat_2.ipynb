```python
import os
import re
from datetime import datetime


def parse_log_files(folder_path, boo=False):
    log_data = []
    # 读取日志文件
    with open(folder_path, 'r', encoding='utf-8') as f:
        for line in f:
            # 找出所有方括号内的数据项
            items = re.findall(r'\[(.*?)\]', line)
            if len(items) < 2:
                continue
            # 提取时间戳
            date = items[0]
            usid = None
            # 遍历所有数据项，查找包含 usid= 的项
            for item in items:
                if item.startswith('usid='):
                    usid = item.split('=')[1]
                    break
            # 保存有效记录
            if usid:
                if '&' in usid:
                    usid = usid.split('&')[0]
                log_data.append({
                    'usid': usid,
                    'date': date
                })
                if boo:
                    print(usid, date, folder_path)
    return log_data
```

```python
file = '/Users/pangmengting/Downloads/发现杭州/20250225统计/web_access_nov_dec_2024_2025/web_access.2024-12-03.txt'
data = parse_log_files(file, True)
len(data)
```
