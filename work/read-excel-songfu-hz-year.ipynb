```python
import pandas as pd

# 读取Excel文件
excel_file = '/Users/pangmengting/Downloads/发现杭州/2025“宋”福杭州年系列活动汇总表-0117.xlsx'  # 将此替换为实际的Excel文件路径
df = pd.read_excel(excel_file, sheet_name=1, skiprows=1)

# 检查列名
print("列名：", df.columns.tolist())

# 清洗列名，去除空格
df.columns = df.columns.str.strip()


# 定义SQL生成函数
def generate_sql(row):
    print(row['活动时间'])

    contact_way = '' if pd.isna(row['其他说明']) else row['其他说明']

    return f"""
        INSERT INTO activity (county_id, type, name, content, address, activity_time, contact_way) 
        VALUES (
            {row['区县']}, 
            4, 
            '{row['项目名称']}', 
            '{row['活动内容']}', 
            '{row['活动地点']}', 
            '{row['活动时间']}', 
            '{contact_way}'
        );
        """


# 生成SQL语句
sql_statements = [generate_sql(row) for _, row in df.iterrows()]

# 保存到文件
with open('songfu.sql', 'w', encoding='utf-8') as file:
    file.writelines(sql_statements)

print("SQL语句已生成并保存到 output.sql 文件中！")
```

```python
import pandas as pd

# 读取Excel文件，从第二行开始作为表头
excel_file = '/Users/pangmengting/Downloads/发现杭州/2025“宋”福杭州年系列活动汇总表-0117.xlsx'  # 将此替换为实际的Excel文件路径
df = pd.read_excel(excel_file, header=1)  # 第二行作为表头

# 定义列名映射关系
column_mapping = {
    '区县': 'county_id',
    '项目名称': 'name',
    '活动内容': 'content',
    '活动地点': 'address',
    '活动时间': 'activity_time',
    '其他说明': 'contact_way',
}

# 重命名DataFrame列
df.rename(columns=column_mapping, inplace=True)


# 定义SQL生成函数
def generate_sql(row):
    return f"""
    INSERT INTO activity (county_id, type, name, content, address, activity_time, contact_way) 
    VALUES (
        {row['county_id']}, 
        4, 
        '{row['name']}', 
        '{row['content']}', 
        '{row['address']}', 
        '{row['activity_time']}', 
        '{row['contact_way']}'
    );
    """


# 生成SQL语句
sql_statements = [generate_sql(row) for _, row in df.iterrows()]

# 保存到文件
with open('songfu.sql', 'w', encoding='utf-8') as file:
    file.writelines(sql_statements)

print("SQL语句已生成并保存到 output.sql 文件中！")
```

```python
import pandas as pd

excel_file = '/Users/pangmengting/Downloads/发现杭州/20250117.xlsx'  # 将此替换为实际的Excel文件路径
df = pd.read_excel(excel_file, header=1)  # 第二行作为表头

# 打印每一行数据（调试用）
print("读取的每一行数据如下：")
for index, row in df.iterrows():
    print(f"第 {index + 1} 行：", row.to_dict())

# 定义列名映射关系
column_mapping = {
    '区县': 'county_id',
    '项目名称': 'name',
    '活动内容': 'content',
    '活动地点': 'address',
    '活动时间': 'activity_time',
    '其他说明': 'contact_way',
}

# 重命名DataFrame列
df.rename(columns=column_mapping, inplace=True)

# 打印重命名后的列名（调试用）
print("\n重命名后的列名：", df.columns.tolist())


# 定义SQL生成函数
def generate_sql(row):
    return f"""
    INSERT INTO activity (county_id, type, name, content, address, activity_time, contact_way) 
    VALUES (
        {row['county_id']}, 
        4, 
        '{row['name']}', 
        '{row['content']}', 
        '{row['address']}', 
        '{row['activity_time']}', 
        '{row['contact_way']}'
    );
    """


# 生成SQL语句并打印（调试用）
print("\n生成的SQL语句如下：")
sql_statements = []
for _, row in df.iterrows():
    sql = generate_sql(row)
    print(sql)  # 打印每条SQL语句
    sql_statements.append(sql)

# 保存到文件
with open('output.sql', 'w', encoding='utf-8') as file:
    file.writelines(sql_statements)

print("\nSQL语句已生成并保存到 output.sql 文件中！")
```
